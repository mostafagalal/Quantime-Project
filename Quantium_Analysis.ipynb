```python
df_all = pd.read_csv("Full_data_cleaned.csv")
```

```python
df_all.head()
```

```python
# Changing the data types from int to String
df_all[['Store_Nbr','Lylty_Card_Nbr','Txn_Id','Prod_Nbr']] = df_all[['Store_Nbr','Lylty_Card_Nbr','Txn_Id','Prod_Nbr']].astype(str)
```

```python
# descriptive analysis for the numeric columns
for i in df_all.columns:
    if np.issubdtype(df_all[i].dtypes , np.number):
        print(f"Column Name = {df_all[i].name} ")
        print(Data_Analysis.gen_stats(df_all[i]))
        print("#"*50)
```

```python
# to devide the package_size column into intervals 
bins_size = 15
bins = np.arange(df_all.Package_Size.min() , df_all.Package_Size.max()+bins_size, bins_size)
df_all['package_bins'] = pd.cut(df_all.Package_Size , bins=bins , right=False)
```

```python
# the most frequent bin 
df_all['package_bins'].value_counts()
```

```python
# New table that contains the most frequent interval from 175 to less than 190 
most_package = df_all[df_all.package_bins == pd.Interval(175.0 , 190.0 , closed = 'left')].copy()
```

```python
# Most Age_Stage to buy the most frequent package from 175.0 , 190.0
most_package.Age_Stage.value_counts()
```

```python
# Most Premium_Customer to buy the most frequent package from 175.0 , 190.0
most_package.Premium_Customer.value_counts()
```

```python
# Most Marital to buy the most frequent package from 175.0 , 190.0
most_package.Marital.value_counts()
```

```python
# grouping 'Age_Stage','Marital' & the most frequent package from 175.0 , 190.0 value_counts
most_package.groupby(['Age_Stage','Marital'])['package_bins'].value_counts().reset_index(name = 'Total').query('Total > 0').sort_values('Total',ascending=False)
```

```python
# Most 5 Brands has sales ?
Data_Analysis.top_n(df_all,['Brand'],'Tot_Sales')
```

```python
# Most 5 Flavors has sales ?
Data_Analysis.top_n(df_all,['Flavor'],'Tot_Sales')
```

```python
# most package size has been sold ?
df_all.groupby('package_bins',observed=True)['Prod_Qty'].sum()\
    .reset_index(name = 'Total Quantity').sort_values('Total Quantity' , ascending = False).reset_index(drop=True).head(5)
```

```python
# most quantity has been sold from which brand ? 
df_all.groupby('Brand')['Prod_Qty'].sum()\
    .reset_index(name='Total Quantity Sold').sort_values('Total Quantity Sold' , ascending = False).reset_index(drop=True).head(5)
Data_Analysis.top_n(df_all,'Brand','Prod_Qty', mode='count', n=5)
```

```python
# most quantity has been sold from which flavour ?
df_all.groupby('Flavor')['Prod_Qty'].sum()\
    .reset_index(name='Total Quantity Sold').sort_values('Total Quantity Sold' , ascending = False).reset_index(drop=True).head(5)
```

```python
# Changing the date column's data type to DATE 
df_all['Date'] = pd.to_datetime(df_all.Date)
```

```python
# monthly sales growth
monthly_sales_growth = df_all.groupby(df_all.Date.dt.month)['Tot_Sales'].sum()
monthly_sales_growth.pct_change()*100
```

```python
# monthly Quantity growth
monthly_quantity_growth = df_all.groupby(df_all.Date.dt.month)['Prod_Qty'].sum()
monthly_quantity_growth.pct_change()*100
```

```python
# which store has the most sales monthly 
Data_Analysis.top_n(df_all,['Store_Nbr'],'Tot_Sales')
```

```python
# sales by store & brand 
Data_Analysis.top_n(df_all,['Store_Nbr','Brand'],'Prod_Qty',n = 10)
```

```python
# Most package bin sold as per quantity on monthly basis 
Monthly_package = df_all.groupby([df_all.Date.dt.month,'package_bins'],observed = True)['Prod_Qty'].sum()\
    .reset_index(name = 'Total Quantity Sold').sort_values( 'Total Quantity Sold' , ascending = False).head(12).reset_index(drop=True).copy()
Monthly_package.sort_values(by = 'Date',ascending = True).reset_index(drop=True)
```
